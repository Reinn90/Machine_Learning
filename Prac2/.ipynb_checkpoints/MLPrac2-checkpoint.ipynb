### What is the general form of linear regression? What is its vector form?
General Form of Linear Regression
Linear regression is a supervised machine learning algorithm used to model the relationship between a dependent variable (target) and one or more independent variables (features). The general form of linear regression for a single feature is given by the equation:

$$\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2x_2 + \dots + \theta_nx_n$$
Where:
- $\hat{y}$ is the **PREDICTED VALUE**
- $n$ is the number of features
- $x_i$ is the $ith$ feature value
- $\theta_0$ is the $jth$ model parameter
- $\theta_0$: term
- $\theta_1 , ... \theta_n$: feature weights

Vector Form of Linear Regression:
In vector form, for a dataset with multiple features, the equation can be represented as:

$$\hat{y} = h_\theta (x) = \theta  x$$
Where:
- $\theta$ is the model's parameter vector
- $x$ is the instance's feature vector, $x_0 = 1$
- $\theta x$ is the dot product of the vectors $\theta$ and $x$
- $h_\theta$ is the **hypothesis function**, using the model parameters $\theta$

### What is the purpose of cost function? What is Mean Square Error (MSE)? What is the normal equation that minimizes the cost function MSE(θ)?

Purpose of Cost Function:
The purpose of a cost function (also called a loss function) in linear regression is to measure how well the chosen model's predictions align with the actual target values. The goal is to minimize the cost function by adjusting the model's parameters (coefficients) to achieve the best possible fit to the data.

Mean Square Error (MSE):
Mean Square Error is a commonly used cost function in linear regression. It measures the average squared difference between the predicted values and the actual target values:

$$MSE(X, h_\theta) = \frac{1}{m}\sum^m_{i=1} (\theta^T x^{(i)} - y^{(i)})^2$$


Where:
- $m$ is the number of data points.
- $\hat{\theta}$: the value of $\theta$ that minimizes the cost function $MSE(\theta)$
- $x^{(i)}$​ is the actual target value for the $i$-th data point.
- $y^{(i)}$​ is the predicted target value for the $i$-th data point.

Normal Equation:
The normal equation is a mathematical approach to finding the optimal coefficients that minimize the cost function (MSE) directly. It is given by the formula:

$$ \hat{\theta} = (X^T X)^{-1}X^Ty$$

- $\hat{\theta}$: the value of $\theta$ that minimizes the cost function $MSE(\theta)$
- $y$ is the vector of target values: $y^{(1)}$ to $y^{(m)}$


### Describe differences between Batch gradient descent, Stochastic gradient descent and Mini-batch gradient descent.

Differences between Gradient Descent Variants:

- Batch Gradient Descent: In batch gradient descent, the entire dataset is used to compute the gradient of the cost function in each iteration. It provides a more accurate estimate of the gradient but can be computationally expensive for large datasets.

- Stochastic Gradient Descent (SGD): In SGD, only one random data point is used to compute the gradient in each iteration. It introduces more randomness but can converge faster and handle large datasets better. However, it can have a lot of variance in the optimization process.

- Mini-Batch Gradient Descent: Mini-batch gradient descent strikes a balance between batch and stochastic approaches. It uses a small random subset (mini-batch) of the data to compute the gradient. This combines the advantages of both methods, offering faster convergence with less noise compared to SGD.

In summary, batch gradient descent provides accurate but slow updates, SGD is fast but noisy, and mini-batch gradient descent is a compromise between the two, often preferred in practice.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(0)

# Simulated dataframe for taxi fare based on kms traveled, the $ on the meter, and the total Paid including tips)
df_sim = pd.DataFrame(
    {
        'Kilometres': [1, 2, 3, 4, 5],
        'Meter': [7.5, 10, 12.5, 15, 17.5],
        'Paid (incl. tips)': [8, 10, 14, 15, 20],
    }
)

df_sim.set_index('Kilometres') # set the first column as the index of the df as it is an increment of 1

# df_sim.set_index('Kilometres')[['Meter']] # show the meter column

# df_sim.set_index('Kilometres')[['Meter', 'Paid (incl. tips)']] # also show the Paid column

```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 5), sharex=False, sharey=False)

df_sim.set_index('Kilometres')['Meter'].plot(title='Meter',\
                                             kind='line', marker='o', zorder=1, ax=axs[0], grid=True)
df_sim.set_index('Kilometres')['Paid (incl. tips)'].plot(title='Paid (incl. tips)', \
                                                         kind='line', marker='o', linestyle='--', ax=axs[1], grid=True)

axs[0].set_xlabel('Kilometres')
axs[1].set_xlabel('Kilometres')

axs[0].set_ylabel('Amount ($)')
axs[1].set_ylabel('Amount ($)')

axs[0].set_xlim((-0.1,5.1))
axs[0].set_ylim((-0.1,20.5))

axs[1].set_xlim((-0.1,5.1))
axs[1].set_ylim((-0.1,20.5))

# axs[0].plot([0,1], [5,7.5], linestyle='--', color='k', zorder=0, alpha=0.6)

fig.show()
```

```python
df_taxi = df_sim
print(df_taxi)  #add print
```

```python
from sklearn.linear_model import LinearRegression

# Initialize and train the model
reg = LinearRegression()
reg.fit(df_taxi[['Kilometres']], df_taxi['Paid (incl. tips)'])
df_taxi['Paid (Predicted)'] = reg.predict(df_taxi[['Kilometres']])


print(
    'Amount Paid = {:.1f} + {:.1f} * Distance'.format(
        reg.intercept_, reg.coef_[0], 
    )
) 
```

```python
# Plot

# Initialise a shared plot for actual and regression line, but dont share axes
fig, ax = plt.subplots(1, 1, figsize=(10, 8), sharex=False, sharey=False)

# Scatterplot of actual values
df_taxi.tail(5).plot(
    label='actual', kind='scatter', x='Kilometres', y='Paid (incl. tips)',
    marker='o', alpha=0.95, ax=ax
)

# Regression line
df_taxi.tail(5).set_index('Kilometres')['Paid (Predicted)'].plot(
    title='Paid (incl. tips)', label='estimated', kind='line', 
    marker='o', linestyle='--', color='k', alpha=0.95, ax=ax
)

# X axis limit values
x_min, x_max = 0.8, 5.2

# Y axis limit values
y_min, y_max = 5, 22

ax.set_xlim((x_min, x_max))
ax.set_ylim((y_min, y_max))

# show the distances between actual and regression points
for x, y1, y2 in df_taxi.tail(5)[
    ['Kilometres', 'Paid (incl. tips)', 'Paid (Predicted)']
].values.tolist():
    ymin, ymax = min(y1, y2), max(y1, y2)
    ymin, ymax = (ymin-y_min)/(y_max-y_min) , (ymax-y_min)/(y_max-y_min)
    ax.axvline(x=x, ymin=ymin, ymax=ymax, color='r')

ax.legend(loc='lower right', fontsize='xx-large')

fig.show()
```

```python
# plot meter line and regression
fig, axs = plt.subplots(1, 2, figsize=(16, 5), sharex=False, sharey=False)

df_taxi.set_index('Kilometres')['Meter'].plot(
    title='Meter', kind='line', 
    marker='o', alpha=0.75, ax=axs[0]
)
df_taxi.set_index('Kilometres')['Paid (incl. tips)'].plot(
    label='actual', kind='line', 
    marker='o', linestyle='--', alpha=0.75, ax=axs[1]
)

df_taxi.set_index('Kilometres')['Paid (Predicted)'].plot(
    title='Paid (incl. tips)', label='estimated', kind='line', 
    marker='o', linestyle='-.', color='k', alpha=0.75, ax=axs[1]
)

axs[0].set_xlabel('Kilometres')
axs[1].set_xlabel('Kilometres')

axs[0].set_ylabel('Amount ($)')
axs[1].set_ylabel('Amount ($)')

axs[0].set_xlim((-0.1,5.1))
axs[0].set_ylim((-0.1,20.5))

axs[1].set_xlim((-0.1,5.1))
axs[1].set_ylim((-0.1,20.5))

axs[1].legend(
    markerscale=0.01, loc='lower right', fontsize='xx-large'
)

fig.show()
```

## 3) For the Stochastic Gradient Descent implementation, try three different learning schedules (e.g. change t1) and compare the results.


### a. t0 = 5, t1 = 50

```python
################
################ Stochastic linear gression 

n_epochs = 50
m=100
t0, t1 = 5, 50 # learning schedule hyperparameters
def learning_schedule(t):
    return t0 / (t + t1)
    
theta = np.random.randn(2,1) # random initialization

for epoch in range(n_epochs):
    for i in range(m): # total of 5000 iterations
        random_index = np.random.randint(m)
        # take a row section of the X and y
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        #calculatethe gradients
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
print("stochastic", theta)
```
