### 3. What is logistic regression? What is it used for?

Logistic regression is a statistical method used for binary classification, which means it is used to predict the probability that an input belongs to one of two possible classes. 

- Ouputs probability margins, between 0 and 1, for estimating membership to a particular class


### 2. Compare linear regression and logistic regression.
1. Purpose:

    Linear Regression: Predicts a continuous numeric value (regression).
   
    Logistic Regression: Estimates the probability of a binary outcome (classification).

3. Outcome:

    Linear Regression: Continuous and unbounded.
   
    Logistic Regression: Probability between 0 and 1.

5. Equation:

    Linear Regression: Simple linear equation (y = mx + b) or multiple linear equation.
   
    Logistic Regression: Uses the logistic function to transform a linear equation.

7. Output Interpretation:

    Linear Regression: Represents the expected numeric value of the outcome.
   
    Logistic Regression: Represents the probability that an instance belongs to a particular class.

9. Loss Function:

    Linear Regression: Typically uses Mean Squared Error (MSE).
   
    Logistic Regression: Uses Log Loss (Cross-Entropy) to measure the difference between predicted probabilities and actual outcomes.

11. Algorithm:

    Linear Regression: Solved using analytical methods or optimization techniques like gradient descent.
    
    Logistic Regression: Optimized using techniques like gradient descent or Newton-Raphson.

13. Multiclass Extension:

    Linear Regression: Generally not used for multiclass classification without modifications.
    
    Logistic Regression: Can be extended to handle multiclass classification using techniques like One-vs-Rest or Softmax Regression.

15. Assumptions:

    Linear Regression: Assumes a linear relationship between predictors and the outcome.
    
    Logistic Regression: Assumes a linear relationship between predictors and the log-odds of the outcome.

17. Application:

    Linear Regression: Predicting stock prices, house prices, temperature, etc.
    
    Logistic Regression: Spam detection, medical diagnosis, credit risk assessment, etc.

19. Evaluation:

    Linear Regression: Often evaluated using metrics like Mean Absolute Error or Root Mean Squared Error.
    
    Logistic Regression: Evaluated using metrics like accuracy, precision, recall, and F1-score.

In summary, while both linear regression and logistic regression are regression techniques, they serve different purposes: linear regression predicts continuous outcomes, while logistic regression predicts probabilities for binary outcomes. The choice between them depends on the nature of the problem and the type of data being analyzed.

### 3. What is the general form of log loss function? Does it have a closed form solution? Is it convex?

The general form of the log loss function (cross-entropy loss) for binary classification is:

$$\text{Log Loss} = -\left[y \cdot \log(p) + (1 - y) \cdot \log(1 - p)\right]$$

Where:
- $y$ is the true binary label (0 or 1) of the instance.
- $p$ is the predicted probability of the positive class (class 1).


Closed-Form Solution
- The log loss function doesn't typically have a closed-form solution in the sense that you can't directly solve for the optimal parameters analytically. However, you can use optimization algorithms like gradient descent, Newton-Raphson, or other numerical optimization techniques to find the optimal parameter values that minimize the log loss.

Convexity
- The log loss function is convex. This means that any local minimum that the optimization algorithm converges to is also a global minimum. Convexity ensures that the optimization process in logistic regression (or other algorithms using the log loss) is more likely to find the optimal solution, as there are no "traps" where the optimization might get stuck.

```python
# -*- coding: utf-8 -*-
"""
Created 2023

@author: Geron, A. (2019). Hands-on Machine Learning with Scikit-Learn, Keras and TensorFlow
"""
import numpy as np

import matplotlib.pyplot as plt

#load data

from sklearn import datasets
iris = datasets.load_iris()  
list(iris.keys())

#['data','target', 'frame', 'target_names', 'DESCR', 'feature_names', 'filename']

X = iris["data"][:, 3:] # petal width
y = (iris["target"] == 2).astype(np.int32)  #  1  if Iris-Virginica, else 0
```

```python
######## training ######################

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X, y)  

#predict: estimated probabilities for flowers with petal widths 
#varying from 0 to 3 cm (Figure 4-23):

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")
# add text labels
plt.text(0.5, 0.85, "Not Iris virginica", fontsize=14, color="b", ha="center")
plt.text(0.5, 0.1, "Iris virginica", fontsize=14, color="g", ha="center")
plt.xlabel("Petal width", fontsize=14)
plt.ylabel("Probability", fontsize=14)

print(log_reg.predict([[1.7], [1.5]]))
#array([1, 0]) 
# 1.7cm petal width is classified as Iris Virginica, 
# 1.5cm petal width is classified as Not Iris Virginica
```

```python
###################### training with two features #############

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.int32) # Iris-Virginica

log_reg = LogisticRegression(solver="lbfgs", C=10**10, random_state=42)
log_reg.fit(X, y)

x0, x1 = np.meshgrid(
        np.linspace(2.9, 7, 500).reshape(-1, 1),
        np.linspace(0.8, 2.7, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]

y_proba = log_reg.predict_proba(X_new)

plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")

zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)


left_right = np.array([2.9, 7])
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]

plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.text(3.5, 1.5, "Not Iris virginica", fontsize=14, color="b", ha="center")
plt.text(6.5, 2.3, "Iris virginica", fontsize=14, color="g", ha="center")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.axis([2.9, 7, 0.8, 2.7])

plt.show()
```

```python
########################softmax for multiclass classification###############

X = iris["data"][:, (2, 3)] # petal length, petal width
y = iris["target"] 

softmax_reg = LogisticRegression(multi_class="multinomial", solver="lbfgs", C=10)

softmax_reg.fit(X, y)

softmax_reg.predict_proba([[5, 2]])

# array([[6.38014896e-07, 5.74929995e-02, 9.42506362e-01]])

softmax_reg.predict([[5, 2]])

#array([2])

x0, x1 = np.meshgrid(
        np.linspace(0, 8, 500).reshape(-1, 1),
        np.linspace(0, 3.5, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]


y_proba = softmax_reg.predict_proba(X_new)
y_predict = softmax_reg.predict(X_new)

from sklearn.metrics import confusion_matrix
y_predict1=softmax_reg.predict(X)
confusion_matrix(y, y_predict1)

plt.figure(figsize=(10, 4))
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="Iris virginica")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="Iris versicolor")
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="Iris setosa")
```

## 3. The files students1.csv and students2.csv, available on vUWS, contain examples about subject results, pass(1)/fail ()0), versus average study time per week.

1) Use scikit-learn to compute a logistic regression model from students1.csv for the
classification.


```python
import pandas as pd

st1 = pd.read_csv('students1.csv')

X = st1.iloc[:,0].values.reshape(-1,1) # training values
y = st1.iloc[:,1].values # target features

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X, y) 
```

2) Use the model to classify the (training) instances in students1.csv. How many of the
examples are misclassified? What are the precision and recall values?


```python
y_pred = log_reg.predict(X) # use the model to predict on the same training values

# Create the confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y, y_pred)

# plot the confusion matrix as a heatmap
cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [False, True])

cm_display.plot()
plt.show() 

# Generate the precision and recall scores
from sklearn.metrics import precision_score, recall_score

ps = precision_score(y, y_pred).round(2)*100
print(f'The precision score for the model is {ps}%')

rs = recall_score(y, y_pred).round(2)*100
print(f'The recall score for the model is {rs}%')
```

3) Use the model to classify the instances in students2.csv. How many of the instances
are misclassfied in this case? What are the precision and recall values?

```python
st2 = pd.read_csv('students2.csv')

X_new = st2.iloc[:,0].values.reshape(-1,1)
y_new = st2.iloc[:,1].values

new_ypred = log_reg.predict(X_new)

cm2 = confusion_matrix(y_new, new_ypred)
cm2_display = ConfusionMatrixDisplay(confusion_matrix = cm2, display_labels = [False, True])

cm2_display.plot()
plt.show() 

ps2 = precision_score(y_new, new_ypred).round(2)*100
print(f'The precision score for the model is {ps2}%')

rs2 = recall_score(y_new, new_ypred).round(2)*100
print(f'The recall score for the model is {rs2}%')
```
